```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

Jensen Shannon Divergence per FP in EU and USA dataset (rows: EU topics, columns: USA topics)

```python
JSD_FP4 = pd.read_pickle('compared_FPs/JSD_FP4')
JSD_FP4 = JSD_FP4[JSD_FP4.columns].astype(float)
sns.set(font_scale=1.5)
plt.subplots(figsize=(15,10))
ax = sns.heatmap(JSD_FP4, cmap="YlGnBu")

# turn the axis label
for item in ax.get_yticklabels():
    item.set_rotation(0)

# save figure
plt.savefig('heatmaps/heatmapFP4.png')
```
